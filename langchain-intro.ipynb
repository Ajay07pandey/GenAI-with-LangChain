# Chat with your PDF

As we are learning LangChain so here we are building a application where we can chat with our uploaded PDFs.

## Langchain - It is a open source framework for devloping application powered by  LLM (Large Language Models)

## Why we are using Langchain:

Ques - Suppose you want to search what are the assumptions we take in the linear regression.

Ans - For searching this one method is **Keyword search** where we can search the pdf  by these words (Assumptions,linear regression,etc), but here it is searching keyword wise so in all the pdf where these words are present it will search,not by its's context or (Meaning), like what the user want to search,so to fulfil that we do **Sematic Search** where it is searching the pdf by knowing it's context.

Below is the workflow diagram of doing it.

**Semantic search** : for doing the semantic search first we will convert all the relevent text chunks in numerical representation and vectors. which is also called **Embeddings**.

**Ques**- How many runs virat kohli scored? - (A Unique Vector)

**Ans** - Now it will convert other 3 relevant text chunks into the vector representation, and now we have 4 vectors one of question and other of 3 relevant text chunks, now for the question vector which vector is closely related with the answer is called the relevent one and give you output.


**Work Flow** 

User will provide PDF, It can has any number of pages (Suppose 1000) now with the help of Embeddings we will convert every page into a vector embedding now we have aroud 1000 vectors in N-Dimensional space. and store all the vectors embedding into the Database.

Now if user ask a query then this will convert it into embedding and with the help of **Semantic Search** find the relevant vector in the database with the help of nearest distance formula, and then it will select the relevant 5 pages and send them to the system query for processing.

Finally the brain will do the NLU and Context aware text generation.

**Brain** = Understand the query + Generating the relevant text.

Breakthrough in 2017 when Transformers and LLM came into existence. which can do NLU and Context aware text Generation.

**Brain = LLM**
